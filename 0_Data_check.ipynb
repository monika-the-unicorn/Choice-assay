```python
# List the files in B://azure/labwho/beecam and look for incomplete files (starting az)
from pathlib import Path

beecam_path = Path("B:/azure/labwho/beecam")
mp4_files = beecam_path.rglob("*.mp4")
print(f"MP4 files: {len(list(mp4_files))}")
```

```python
# Test that all mp4 files fit the filename pattern:
# [right|left]_02-12-2024_01-01-10.mp4
import re

pattern = re.compile(r"^(right|left)_\d{2}-\d{2}-\d{4}_\d{2}-\d{2}-\d{2}\.mp4$")
invalid_files = []
for mp4_file in mp4_files:
    if not pattern.match(mp4_file.name):
        invalid_files.append(mp4_file)
if invalid_files:
    print("Invalid filenames found:")
    for invalid_file in invalid_files:
        print(invalid_file)
```

```python
# Use Azure python modules to check all the files in the labwho/beecam Azure container and find any that are over 100Mb

from azure.storage.blob import BlobServiceClient

connect_str = "BlobEndpoint=https://labwho.blob.core.windows.net/;QueueEndpoint=https://labwho.queue.core.windows.net/;FileEndpoint=https://labwho.file.core.windows.net/;TableEndpoint=https://labwho.table.core.windows.net/;SharedAccessSignature=sv=2024-11-04&ss=bfqt&srt=sco&sp=rwdlacupiytfx&se=2027-08-07T22:04:39Z&st=2025-08-07T13:49:39Z&spr=https&sig=Atern21TyIxuZ5pmj%2BMlrcTv9ZQISkqEUnJwubYO84w%3D"
blob_service_client = BlobServiceClient.from_connection_string(connect_str)
container_client = blob_service_client.get_container_client("beecam")
large_files = []
i = 0
for blob in container_client.list_blobs():
    i += 1
    if blob.size > 10 * 1024 * 1024:  # 100 MB
        large_files.append((blob.name, blob.size))
    if i % 1000 == 0:
        print(f"Checked {i} blobs...")

print(f"Found {len(large_files)} files over 10 MB:")
for name, size in large_files:
    print(f"{name}: {size / (1024 * 1024):.2f} MB")
total_large_size = sum(size for name, size in large_files)
print(f"Total size of large files: {total_large_size / (1024 * 1024):.2f} MB")
```
